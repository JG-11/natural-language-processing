# Setup

```
import nltk

nltk.download('cess_esp') # download a corpus. More samples: http://www.nltk.org/nltk_data/
```

# Regular Expressions

```
import re # more information: https://docs.python.org/3/library/re.html

corpus = nltk.corpus.cess_esp.sents()
print(corpus)
print(len(corpus))
```

```
flatten = [w for l in corpus for w in l] # list comprehension
print(flatten)
print(len(flatten))

print(flatten[:10]) # the first ten words
```

```
# re.search(pattern, string)
arr = [w for w in flatten if re.search('es', w)] # metacharacter is a substring that defines a search pattern
print(arr[:5])
```

```
arr = [w for w in flatten if re.search('es$', w)] # $ matches the end of the string
print(arr[:5])
```

```
arr = [w for w in flatten if re.search('^es', w)] # ^ matches the start of the string
print(arr[:5])
```

```
# Set of characters [a-z] [ghi]
arr = [w for w in flatten if re.search('^[ghi]', w)] # a string starting with g, h or i
print(arr[:10])
```

```
"""
  * zero or more times
  + one or more times
"""

arr = [w for w in flatten if re.search('^(no)*', w)]
print(arr[:10])

arr = [w for w in flatten if re.search('^(no)+', w)]
print(arr[:10])
```
