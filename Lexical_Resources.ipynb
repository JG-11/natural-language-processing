# Lexical resources (lexicons)

These are collections of words or phrases that have associated tags or meta-data of some type (e.g. POS tags, gramatical meanings).

P.S. POS means Part of Speech, also called gramatical tagging or words tagging by categories (adjectives, verbs, adverbs, nouns, prepositions, etc.).

```
# if a text has a lot of stopwords, the text is not so big as I thought initially
def stopwords_percentage(text):
  stop_words = stopwords.words('english')
  content = [w for w in text if w.lower() not in stop_words]

  return 1 - (len(content) / len(text1))
```

```
stopwords_percentage(text1)
```
